
Read Google Mobility Report File 



```python
import pandas as pd

mobility_url = 'https://www.gstatic.com/covid19/mobility/Global_Mobility_Report.csv'

mobility = pd.read_csv(mobility_url,
                     parse_dates=['date'],
                     index_col=['country_region', 'date'],
                     squeeze=True).sort_index()
```

```python

#Filter out subregions 

mobility = mobility[mobility['sub_region_1'].isnull()]

```

```python
mobility
```

Now we read the Rt file 

```python


rt_uri = 'data/rt-latest.csv'

rt = pd.read_csv(rt_uri,
                     usecols=[0,1,2,3,4,5,6],
                     parse_dates=['date'],
                     #index_col=['location', 'date'],
                     squeeze=True)


```

```python
mergedRtMobility  = rt.merge(mobility,left_on=['location','date'],right_on=['country_region','date'])


mergedRtMobility.to_csv('data/merged-rt-mobility.csv')
```
